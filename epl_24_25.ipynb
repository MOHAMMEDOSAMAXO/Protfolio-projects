```python
def team_color(df):
    color =[]
    for val in df:
        if val =='Arsenal': color.append('#EF0107')
        elif val =='Bournemouth': color.append("#DA291C")
        elif val =='Chelsea': color.append("#034694")
        elif val =='Brighton': color.append("#0057B8")
        elif val =='Wolves': color.append("#FDB913")
        elif val =='Crystal Palace': color.append("#1B458F")
        elif val =="Nott'ham Forest": color.append("#E53233")
        elif val =='Brentford': color.append("#D20000")
        elif val =='Manchester City': color.append("#6CABDD")
        elif val =='Southampton': color.append("#D71920")
        elif val =='Everton': color.append("#003399")
        elif val =='Liverpool': color.append("#c8102E")
        elif val =='Newcastle Utd': color.append("#241F20")
        elif val =='Leicester City': color.append("#003090")
        elif val =='West Ham': color.append("#7A263A")
        elif val =='Fulham': color.append("#000000")
        elif val =='Aston Villa': color.append("#95bfe5")
        elif val =='Tottenham': color.append("#132257")
        elif val =='Ipswich Town': color.append("#002d62")
        elif val =='Manchester Utd': color.append("#DA291C")
    return color
```

```python
df_epl['MinsPerMatch'] = (df_epl['Min']/df_epl['MP']).astype(int)
df_epl['GoalsPreMatch']= (df_epl['Gls']/df_epl['MP']).astype(float)
df_epl.head()
```

```python
df_epl['Nation']=(df_epl['Nation'].astype(str)).apply(lambda x:(x.split(" ")[-1]))
```

```python
df_epl['Nation'].unique()
```

```python
nation = df_epl.groupby("Nation").size().sort_values(ascending=False)
nation.head(10).plot(kind='bar',title='Players Nationality')
```

```python
df_epl['Squad'].value_counts().nlargest(5).plot(kind='bar',title='Clubs with the largest number of players ')
```

```python
df_squad_min =df_epl['Squad'].value_counts().nsmallest(5)
df_squad_min.plot(kind='bar',title='Clubs with the least number of players')
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=60)
sns.boxplot(df_epl,x='Squad',y='Age',palette=team_color(df_epl['Squad']))
plt.tight_layout()
```

```python
Goals = pd.DataFrame(df_epl.groupby("Squad",as_index=False)['Gls'].sum())
plt.figure(figsize=(12,6))
plt.xticks(rotation=60)
top_teams_scorer =Goals.sort_values(by='Gls',ascending=False)
cl=team_color(top_teams_scorer['Squad'])
sns.barplot(data = top_teams_scorer,x='Squad',y='Gls',palette=cl)
plt.title("Goals Scored per team")
plt.tight_layout()
```

```python
player_goals = pd.DataFrame(df_epl.groupby("Player",as_index=False)['Gls'].sum())
plt.figure(figsize=(12,6))
plt.xticks(rotation=60)
sns.set_theme(style='dark')
merged = player_goals.merge(df_epl[['Player','Squad']],left_on='Player',right_on='Player',how='left')
top_10_players =merged.sort_values(by='Gls',ascending=False).head(10)
sns.barplot(data=top_10_players ,x= 'Player',y='Gls',palette=team_color(top_10_players['Squad']))
plt.title("Top 10 scorers ")
plt.tight_layout()
```

```python
player_ast = pd.DataFrame(df_epl.groupby("Player",as_index=False)['Ast'].sum())
plt.figure(figsize=(12,6))
plt.xticks(rotation=60)
merged = player_ast.merge(df_epl[['Player','Squad']],left_on='Player',right_on='Player',how='left')
top_10_players_ast =merged.sort_values(by='Ast',ascending=False).head(10)
cl = team_color(top_10_players_ast['Squad'])
sns.barplot(top_10_players_ast,x='Player',y='Ast',palette=cl)
plt.title("Top 10 Assits")

plt.tight_layout()
```
